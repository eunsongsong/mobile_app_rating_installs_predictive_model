```python
playstore_data['Rating'].describe()
```

이제 rating에서의 결측치만 남아있었죠? 그것은 0으로 대체합니다!

```python
playstore_data[playstore_data['Rating'] == 0].head()
```

```python
print(playstore_data.describe())
```

```python
rating = playstore_data[playstore_data['Rating'] != 0 ]
```

rating이 0이 아닌 것들로만 추려서 다시 한번 describe 해봅니다!

```python
print(rating.describe())
```

```python
sns.kdeplot(shade=True,data=rating['Rating'])
```

이것은 rating의 density estimates를 보여주는 것인데 4점대 초반이 제일 많은 것을 알 수 있네요!^^<br/>
shade=True를 통하여 그래프 밑부분을 색칠하여 줄 수 있습니다!

```python
from scipy.stats import kurtosis, skew

print( 'excess kurtosis of  distribution : {}'.format( kurtosis(rating['Rating']) ))
print( 'skewness of distribution: {}'.format( skew(rating['Rating']) ))
```

여기서 kurtosis는 첨도를 말하는 것인데 첨도가 높을수록 확률분포가 뾰족하다는 소리입니다!<br/>
즉, 첨도가 높을수록 데이터의 집중도가 높다는 것을 의미합니다!<br/>
skew는 왜도를 의미합니다! 왜도가 높거나 낮을수록 평균에서 치우쳐 있다는 것을 말해요오!

```python
list_1 = ['Category', 'Installs', 'Type', 'Content Rating']
```

```python
def bar_plot(x,y,y_label,x_label,title,color,ax):
    # plt.figure(figsize=(10,5))
    bar = sns.barplot(x = x,y=y,ax=ax,orient='h')
    plt.ylabel(y_label)
    plt.xlabel(x_label)
    plt.title(title)
    for i, v in enumerate(x.values):
        ax.text(v + 3, i+ .25, str(v), color='black', fontweight='bold')
    return bar
```

```python
fig = plt.figure(figsize=(14,18))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
i = 1
for names in list_1:
    ax1 = fig.add_subplot(2, 2, i)
    df2 = playstore_data[names].value_counts()
    df2 = df2.reset_index()
    bar_plot(x = df2[names],y = df2['index'],y_label = 'Freq',title = 'Bar Chart On {}'.format(names),color='red',ax=ax1,x_label=names)
    i += 1
```

```python
def bar_plot(x,y,y_label,x_label,title,color,ax=None):
    plt.figure(figsize=(5,8))
    bar = sns.barplot(x = x,y=y,orient='h')
    plt.ylabel(y_label)
    plt.xlabel(x_label)
    plt.title(title)
    for i, v in enumerate(x.values):
        bar.text(v + 3, i + .25, str(v), color='black', fontweight='bold')
    return bar
df2 = playstore_data['Genres'].value_counts()
df2 = df2.reset_index()
df2 = df2[df2['Genres'] > 100]
bar_plot(x = df2['Genres'],y = df2['index'],y_label = 'Freq',title = 'Bar Chart On Gerner',color='red',x_label='Genres')
```

category, type, installs, contents rating의 분포를 살펴보았읍니다!

그리고 이제는 installs의 값을 숫자로 바꾸어보겠습니당!

```python
new_col_install = playstore_data['Installs'].apply(lambda a : a.replace('+','').replace(',',''))

def add_list(x):
    sum=' '
    for i in range(0,len(x)):
        sum+=x[i]
    return int(sum)  

playstore_data['Installs'] = new_col_install.apply(lambda a : add_list(a)) 
```

```python
playstore_data
```

이렇게 보시면 installs도 숫자 값으로 바뀐 것을 알 수 있습니다!

이제는 price도 문자형에서 숫자형으로 바꿔보도록 하겠습니다!

```python
def remove_curr(x):
    if x !='0':
        x=x.split('$')[1]
    return float(x)   

playstore_data['Price'] = playstore_data['Price'].apply(lambda a : remove_curr(a))  
playstore_data.head(5)
```
